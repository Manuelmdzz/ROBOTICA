```python
def get_frames(filename):
    video = cv2.VideoCapture(filename)
    while video.isOpened():
        ret, frame = video.read()
        if ret:
            yield frame
        else:
            break
        
    video.release()
    yield None
```

## Task 2: Retrieve and display video frames

```python
for f in get_frames(VFILE):
    if f is None:
        break
    cv2.imshow("frame", f)
    if cv2.waitKey(10) == 27:
        break
cv2.destroyAllWindows()
cv2.waitKey(1)

```

## Task 2: Define function to get a single video frame

```python
# Completar la función para obtener un fotograma específico por índice
def get_frame(filename,index):
    counter = 0    
    video = cv2.VideoCapture(filename)
    while video.isOpened():
        ret, frame = video.read()
        if ret:
            if counter == index:
                return frame
            counter += 1
        else:
            break
        
    video.release()
    return None
```

## Task 4: Drawing a circle

```python
frame = get_frame(VFILE, 200)
cv2.circle(frame, 
           center = (200, 200),
           radius= 50,
           color= (0, 0, 255), 
           thickness=10
           )
cv2.rectangle(frame,
              pt1 = (300, 50),
              pt2 = (500, 150),
              color = (255, 0, 0),
              thickness=2
              )
fix_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(fix_frame)
```

## Task 4: Processing an entire video

```python
counter = 0
for frame in get_frames(VFILE):
    if frame is None: 
        break
    cv2.putText(frame,
                text = str(counter),
                org = (50, 50),
                fontFace = cv2.FONT_HERSHEY_SIMPLEX,
                fontScale = 1,
                color = (0, 255, 0), #  BGR
                thickness=3
    )
    cv2.circle(frame, 
               center = (70, 40),
               radius= 70,
               color= (0, 0, 255), 
               thickness=2,
               )
    cv2.ellipse(frame,
                center = (70, 40),
                axes = (100, 50),
                angle = counter,
                startAngle = 0,
                endAngle = 360,
                color = (255, 0, 0),
                thickness = 3
    )
    cv2.imshow("frame", frame)
    if cv2.waitKey(10) == 27:
        break
    counter += 1

cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

VFILE = "video/hot_rod.mp4"

# --- Parameters ---
total_frames = 3597
skip_frames = total_frames // 15   # 15 frames evenly spread
print(f"Skip frames: {skip_frames}")

# --- Open video ---
cap = cv2.VideoCapture(VFILE)
frames = []

for i in range(0, total_frames, skip_frames):
    cap.set(cv2.CAP_PROP_POS_FRAMES, i)   # jump to frame
    ret, frame = cap.read()
    if not ret:
        break
    # Convert BGR → RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frames.append(frame)

cap.release()

# --- Collage: arrange in grid ---
rows = 3
cols = 5
fig, axes = plt.subplots(rows, cols, figsize=(15, 9))

for ax, img in zip(axes.flatten(), frames):
    ax.imshow(img)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
